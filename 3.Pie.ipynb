```python
from  matplotlib import pyplot as plt
plt.style.use( 'tableau-colorblind10')
slices = [59219, 55466, 47544, 364437]
explode=[0,0,0,0.1]
colors=["#008fd5","#fc4f30","#e5ae37","#6d904f"]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python']
plt.pie(slices,labels=labels,colors=colors,wedgeprops={'edgecolor':'black','linewidth': 2},explode=explode,shadow=True,
        autopct="%1.1f%%",
       startangle=90,radius=0.5,labeldistance=1.2,frame=True)
plt.title("pie chart")

plt.show()
```

```python
from  matplotlib import pyplot as plt
plt.style.use( 'tableau-colorblind10')
slices = [59219, 55466, 47544, 364437]
explode=[0,0,0,0.1]
colors=["#008fd5","#fc4f30","#e5ae37","#6d904f"]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python']
plt.pie(slices,labels=labels,colors=colors,wedgeprops={'edgecolor':'black','linewidth': 3},explode=explode,shadow=True,
        autopct="%1.1f%%",
       startangle=90,radius=0.5,labeldistance=1.2,frame=True,center=(.2,.3))
plt.title("pie chart")

plt.show()
```

```python
from  matplotlib import pyplot as plt
plt.style.use( 'tableau-colorblind10')
slices = [59219, 55466, 47544, 364437]
explode=[0,0,0,0.1]
colors=["#008fd5","#fc4f30","#e5ae37","#6d904f"]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python']
plt.pie(slices,labels=labels,colors=colors,wedgeprops={'edgecolor':'black','linewidth': 3,"linestyle":"--"},
        explode=explode,shadow=True,autopct="%1.1f%%",
       startangle=10,radius=0.5,labeldistance=1.2,frame=True,center=(.2,.3),counterclock=False)
plt.title("pie chart")

plt.show()
```
