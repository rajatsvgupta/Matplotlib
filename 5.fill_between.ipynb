```python
from matplotlib import pyplot as plt
import pandas as pd
data=pd.read_csv("csv_for_fill_in.csv")

age=data["Age"]
dev=data["All_Devs"]
python=data['Python']
java=data['JavaScript']
plt.plot(age,dev)
plt.plot(age,python)
plt.plot(age,java)
avg=57287
plt.fill_between(age,dev,python, where =(dev>python), alpha=0.4,interpolate=True,color="red",label="below")
plt.fill_between(age,dev,python, where =(dev<python), alpha=0.4,interpolate=True,color="black",label="above")
plt.legend()
plt.show()
```
